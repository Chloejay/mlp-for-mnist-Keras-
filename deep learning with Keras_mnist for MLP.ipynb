```python
from __future__ import print_function 
import numpy as np 
from keras.datasets import mnist 
from keras.models import Sequential 
from keras.layers.core import Dense, Activation 
from keras.optimizers import SGD 
from keras.utils import np_utils 
np.random.seed(1234) 
```

```python
NB_EPOCH = 200 #how many iteration that model needs to run during the whole model processing 
BATCH_SIZE = 128 
VERBOSE = 1 
NB_CLASSES = 10 # number of outputs = number of digits, the number of the multi-category labels 
OPTIMIZER = SGD() # SGD optimizer, for the loss and crossentropy, intergate with the learning rate 
N_HIDDEN = 128  
VALIDATION_SPLIT=0.2 # how much TRAIN is reserved for VALIDATION
# data: shuffled and split between train and test sets#
(X_train, y_train), (X_test, y_test) = mnist.load_data()
#X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784
RESHAPED = 784 #image size id 28*28  
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32') #use the datatype as float32 for the GPU computation 
X_test = X_test.astype('float32') 
# normalize#   
X_train /= 255
X_test /= 255 
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, NB_CLASSES) #np_utils 
Y_test = np_utils.to_categorical(y_test, NB_CLASSES) 
```

```python
print(X_train.shape, y_train.shape)   
```

```python
model = Sequential() # the basic model in the Keras 
model.add(Dense(NB_CLASSES, input_shape=(RESHAPED,)))
#model.add(Activation='relu')
model.add(Activation('softmax'))
model.summary()   
```

```python
#model.compile(loss='mean_squared_error', optimizer='sgd') 
```

```python
model.compile(loss='categorical_crossentropy', optimizer=OPTIMIZER, metrics=['accuracy']) 
#categorical_crossentropy for the multiclass loss function, objective function in the Keras 
```

```python
test= model.fit(X_train, Y_train, batch_size= BATCH_SIZE, epochs= NB_EPOCH,
               verbose= VERBOSE, validation_split= VALIDATION_SPLIT) 
```

```python
#evaluate the dataset once the model is trained 
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)  
```

```python
print("Test score:", score[0])
print("Test accuracy", score[1]) 
```

```python
from keras.layers.core import Dense, Dropout, Activation 
#use the dropout to improve the performance 
NB_EPOCH= 250 
BATCH_SIZE= 128 
VERBOSE=1 
NB_CLASSES=10 
OPTIMIZER= SGD() 
N_HIDDEN= 128
VALIDATION_SPLIT=0.2
DROPOUT=0.3 
(X_train, y_train), (X_test, y_test)=mnist.load_data() 
```

```python
RESHAPED = 784
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED) 
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
# normalize
X_train /= 255
X_test /= 255
# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, NB_CLASSES)
Y_test = np_utils.to_categorical(y_test, NB_CLASSES) 
# M_HIDDEN hidden layers 10 outputs
model = Sequential()
model.add(Dense(N_HIDDEN, input_shape=(RESHAPED,)))

model.add(Activation('relu')) 
model.add(Dropout(DROPOUT))
model.add(Dense(N_HIDDEN))

model.add(Activation('relu'))
model.add(Dropout(DROPOUT))

model.add(Dense(NB_CLASSES))
model.add(Activation('softmax'))
model.summary()

model.compile(loss='categorical_crossentropy',optimizer=OPTIMIZER, metrics=['accuracy']) 
test = model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=NB_EPOCH,
verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("Test score:", score[0])
print('Test accuracy:', score[1])   
```
